```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv('book.csv')   #read the data 
print(test)
```

```python
boxplot = test.boxplot(figsize = (10,20), column = 'Total_Bill')  #boxplot for total bill
boxplot.plot()
plt.show()
```

```python
subset = test[['Tips', 'Total_Bill']]
scatter = subset.plot(x = 'Tips', y = 'Total_Bill', kind='scatter', c= 'green');
scatter.set_title('Tips vs Total Bill');
# export the pandas chart to an image
scatter.figure.savefig('Tips_vs_TotalBill.png')  # scatter plot of tip vs total bill

plt.show()
```

```python
list = []
for ind in test.index:
    
 value = 0.18-0.01*(test['Size'][ind])
 list.append(value)

print(list)   #calculte tip rate using the equation
```

```python
test['tip_rate'] = list
```

```python
x1 = []


for ind in test.index:
    val = test['tip_rate'][ind]
    x1.append(val)                  #collect all values of tip_rate
```

```python
y1 = []
for ind in test.index:
    val = test['Size'][ind]
    y1.append(val)                  #3collect all values of Size
```

```python
subset = test[['tip_rate', 'Size']]
scatter = subset.plot(x = 'tip_rate', y = 'Size', kind='scatter', c= 'green')
scatter.set_title('tip_rate vs Size of Party')
slope, intercept = np.polyfit(x1,y1,1)
ab = [slope*i + intercept for i in x1]

plt.plot(x1,y1,'--')
plt.plot(x1,ab,'b')
plt.title('tip_rate')
plt.show()

# export the pandas chart to an image
scatter.figure.savefig('tip_rate_vs_Size.png')   #Scatter plot of tip_rate vs party_size
```
